```
df = pd.read_csv("/work/Results/Motive/Processed/motive_processed_july4.csv")
df
```

```


sns.set(style="whitegrid")
order = ['AI with caring motives', 'AI with no motives', 'AI with manipulative motives']

my_pal = {"AI with no motives": "#ababab", "AI with manipulative motives": "#d16969", "AI with caring motives":"#87c771"}
ax = sns.countplot(x='assigned_group',data=df,  palette=my_pal, capsize=.1)
plt.ylim(1, 80)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
#plt.savefig('example_non-hue_outside.png', dpi=300, bbox_inches='tight')

```

```
sns.set(style="whitegrid")
order = ['AI with caring motives', 'AI with no motives', 'AI with manipulative motives']

my_pal = {"Caring motives": "#87c771", "Manipulative/malicious motives": "#d16969", "No motive":"#ababab", "Other":"#c7c7c7",}
ax = sns.countplot(x='From your own experience, what do you think the motive of the agent was?  Selected Choice',data=df,  palette=my_pal, capsize=.1)
plt.ylim(1, 80)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
#plt.savefig('example_non-hue_outside.png', dpi=300, bbox_inches='tight')

```

```
sns.set(style="whitegrid")
order = ["18-24 years old", "25-34 years old","35-44 years old","45-54 years old","55-64 years old","65+ years old"]
sns.set_palette("Greys_r")
ax = sns.countplot(x='How old are you?',data=df,order=order, capsize=.1)
plt.ylim(1, 80)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
#plt.savefig('example_non-hue_outside.png', dpi=300, bbox_inches='tight')

```

```
sns.set(style="whitegrid")
sns.set_palette("Greys_r")
ax = sns.countplot(x='What gender do you identify as?  Selected Choice',data=df, capsize=.1)

ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
#plt.savefig('example_non-hue_outside.png', dpi=300, bbox_inches='tight')

```
